```python
import pandas as pd
import numpy as np

df = pd.read_csv('dataset.csv')
df
```

```python
# Fill NaNs with empty string
df.fillna('', inplace=True)

# Combine all symptom columns into a single list for easier processing
symptom_cols = [f'Symptom_{i}' for i in range(1, 18)]
df[symptom_cols] = df[symptom_cols].apply(lambda x: x.str.strip().str.lower())

# Preview
df.head()
```

```python
# Load symptom severity
severity_df = pd.read_csv("symptom-severity.csv")
all_symptoms = sorted(severity_df['Symptom'].str.strip().str.lower().unique())

# Create one-hot columns
for symptom in all_symptoms:
    df[symptom] = df[symptom_cols].apply(lambda x: symptom in x.values, axis=1).astype(int)

```

```python
df[all_symptoms]
```

```python
# Features are the symptom presence columns
X = df[all_symptoms]

# Target
y = df['Disease']
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Encode target labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
def predict_disease(symptom_list):
    input_data = [1 if s in symptom_list else 0 for s in all_symptoms]
    pred = model.predict([input_data])[0]
    return le.inverse_transform([pred])[0]
```

```python
user_symptoms = ['itching', 'skin_rash', 'nodal_skin_eruptions']
print("Predicted Disease:", predict_disease(user_symptoms))
```

```python
import joblib


joblib.dump(model, "disease_model.pkl")
joblib.dump(le, "label_encoder.pkl")
joblib.dump(all_symptoms, "symptoms_list.pkl")
```
